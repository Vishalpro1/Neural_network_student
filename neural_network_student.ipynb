```python
data = pd.read_csv("/kaggle/input/student-performance/StudentPerformance.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
# change categorical data into numerical data
from sklearn.preprocessing import LabelEncoder
```

```python
a = LabelEncoder()
data["Extracurricular Activities"] = a.fit_transform(data["Extracurricular Activities"])
data
```

```python
x = data[["Hours Studied",'Previous Scores','Extracurricular Activities','Sleep Hours','Sample Question Papers Practiced']]
```

```python
y =data[['Performance Index']]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size = 0.2 , random_state = 45)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(16,activation = 'relu',input_shape = (5,)))
model.add(Dense(16,activation = 'relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer = 'adam',loss = 'mean_squared_error',metrics=['mae'])
```

```python
model.fit(x_train,y_train,epochs = 100,batch_size = 16,verbose=1)
```

```python
loss,mae = model.evaluate(x_test,y_test)
print("mean absolute error",mae)
```

```python
import numpy as np

new_student = np.array([[5, 52, 1,5,2]])  # hours, previous score, Extracurricular Activities,Sleep Hours,Sample Question Papers Practiced
new_student = sc.transform(new_student)

prediction = model.predict(new_student)
print("Predicted Performance Index:", prediction[0][0])



```
